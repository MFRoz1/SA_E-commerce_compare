```python
import re
def  clean_text_s(data_shopee, text_field, new_text_field_name):
    data_shopee[new_text_field_name] = data_shopee[text_field].str.lower()
    data_shopee[new_text_field_name] = data_shopee[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))
    # remove numbers
    data_shopee[new_text_field_name] = data_shopee[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem))
    return data_shopee
```

```python
data_shopee['text_clean'] = data_shopee['content'].str.lower()
data_shopee['text_clean']
data_clean_shopee = clean_text_s(data_shopee, 'content', 'text_clean')
data_clean_shopee.head(10)
```

###Stopword Removal

```python
# Download stopwords untuk bahasa Indonesia
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))
data_clean_shopee['text_StopWord'] = data_clean_shopee['text_clean'].apply(lambda x:' '.join([word for word in x.split() if word not in (stop_words)]))
data_clean_shopee.head(10)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
data_clean_shopee['text_tokens'] = data_clean_shopee['text_StopWord'].apply(lambda x: word_tokenize(x))
data_clean_shopee.head()
```

```python
def get_main_topic(text):
    # Melakukan preprocessing teks
    processed_text = praproses(text)

    # Melakukan analisis teks untuk menentukan topik utama
    # Anda dapat menggunakan metode seperti TF-IDF, LDA, atau model berbasis deep learning seperti BERT

    # Contoh menggunakan metode TF-IDF
    # Anda perlu mengimpor library yang diperlukan
    from sklearn.feature_extraction.text import TfidfVectorizer

    # Menginisialisasi objek TF-IDF
    tfidf_vectorizer = TfidfVectorizer()

    # Mengubah teks menjadi vektor TF-IDF
    tfidf_matrix = tfidf_vectorizer.fit_transform([processed_text])

    # Mengambil kata-kata dengan bobot tertinggi
    top_words = tfidf_vectorizer.get_feature_names_out()

    # Mengembalikan kata-kata dengan bobot tertinggi sebagai topik utama
    main_topic = top_words[tfidf_matrix.toarray()[0].argmax()]

    return main_topic
```
